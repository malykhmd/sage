## Регулярные пучки

```python
q=2

def regular_bundles():
    L=[]
    var('x')
    K=GF(q)[x]
    for A in MatrixSpace(GF(q),3,3):
        for B in MatrixSpace(GF(q),3,3):
            f=K((A+K(x)*B).det())
            if f.degree()==3 and f.is_prime():
                L.append([A,B])
    return L
```

```python
L=regular_bundles()
```

```python
A,B=L[0]
```

## Преобразование $(A,B) \to T$

```python
q=2
load('tex/Lectures-2025-corr/sage/gauss.sage')

def point_to_number(p):
    P=ProjectiveSpace(2,GF(2))
    return list(P).index(p)+1

def number_to_point(n):
    P=ProjectiveSpace(2,GF(2))
    return list(P)[n-1]

def bileniar_form(A,p,pp):
    return matrix(1,3,list(pp))*A*matrix(3,1,list(p))

def rosanes_to_permutation(A,B):
    ans=[]
    for p in ProjectiveSpace(2,GF(q)):
        for pp in ProjectiveSpace(2,GF(q)):
            if bileniar_form(A,p,pp)==0 % q and bileniar_form(B,p,pp)==0 % q:
                ans.append(point_to_number(pp))
    return SymmetricGroup(len(ans))(ans)
```

```python
def permutation_to_rosanes(t):
    a=var(['a'+str(i) for i in range(9)])
    K=GF(q)[a]
    A=matrix(K,3,3,a)
    eqs=[]
    for p in ProjectiveSpace(2,GF(q)):
        pp=number_to_point(t(point_to_number(p)))
        eqs= eqs + bileniar_form(A,p,pp).list()
    T=triangulation(eqs)
    D=tsolve(T)

    if len(T)==7:
        Vars=[K(aa) for aa in a if K(aa) not in D.keys()]
        A=A.subs(D)
        A1=A.subs(dict([[aa,ZZ(aa==Vars[0])] for aa in Vars]))
        A2=A.subs(dict([[aa,ZZ(aa==Vars[1])] for aa in Vars]))
        return matrix(GF(q),3,3,A1.list()), matrix(GF(q),3,3,A2.list())
    else:
        return A.subs(D)
```
